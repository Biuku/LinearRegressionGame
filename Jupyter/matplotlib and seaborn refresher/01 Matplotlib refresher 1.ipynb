```python
import matplotlib.pyplot as plt
import numpy as np
```

## 1.0 Basic example of plot

```python
x = np.linspace(0, 100, 20)
y = x**2
```

```python
plt.plot(x, y, 'r')

plt.xlabel("X")
plt.ylabel("Y")

plt.show()
```

## 2.0 figure() -- a more OOP approach

```python
fig = plt.figure()

axes = fig.add_axes([0.1, 0.1, 0.8, 0.8]) # left, bottom, width, height (%)

axes.plot(x, y, 'b')
```

```python
fig = plt.figure()

axis1 = fig.add_axes([0.1, 0.1, 1, 0.8])
axis1.plot(x, y, 'b')

axis2 = fig.add_axes([0.2, 0.5, 0.4, 0.3])
axis2.plot(y, x, 'r')
```

## 3.0 Subplots

```python
fig, axes = plt.subplots(figsize=(12, 3))
```

```python
fig, axes = plt.subplots(figsize=(12, 3))

axes.plot(x, y, 'r')
```

## 4.0 Plot range

```python
fig, axes = plt.subplots(1, 2, figsize=(12, 4))  ## rows, cols

axes[0].plot(x, x**2, x, x**3)
axes[0].set_title("Default")


axes[1].plot(x, x**2, x, x**3)
axes[1].set_ylim([-3000, 5000])
axes[1].set_xlim([-10, 50])

axes[1].set_title("Custom range")


                  

```
