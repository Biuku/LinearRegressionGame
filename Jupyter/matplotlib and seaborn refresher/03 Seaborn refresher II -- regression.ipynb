### Basically just lmplot -- linear model plot

```python
import seaborn as sns
import pandas as pd
import numpy as np

sns.set()
```

## 1. Load and explore dataset

```python
df = sns.load_dataset('tips')
```

```python
df.head(3)
```

```python
df.describe()
```

```python
df.columns
```

```python
bill = 'total_bill'
tip = 'tip'
sex = 'sex'
size = 'size'
day = 'day'
time = 'time'
smoker = 'smoker'
```

### 1.1 jointplot()

```python
sns.jointplot(x= bill, y = tip, data = df, kind = 'reg')
```

## 2. Basic lmplot

```python
sns.lmplot(x = 'total_bill', y = 'tip', data = df)
```

## 2. Change hue

```python
sns.lmplot(bill, tip, df, hue = 'sex')
```

## 3. Change colour pallette

```python
sns.lmplot(bill, tip, df, hue=sex, palette = 'coolwarm')
```

## 4. Laying out multiple plots

```python
sns.lmplot(bill, tip, df, col=sex)
```

## 5. Aspect and size 

```python
sns.lmplot(bill, tip, df, col = day, hue=sex, palette ='coolwarm', aspect =0.6, size = 8)
```
